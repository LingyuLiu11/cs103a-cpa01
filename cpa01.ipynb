```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data/data.csv')
```

```python
df2 = df[df['STATUS']=='F']
df2
df2.plot(x='Period', y='Data_value')
plt.xlabel('Period')
plt.ylabel('Data_value')
plt.title('value in years of status F')
plt.grid()
plt.legend()
```

```python

table = pd.pivot_table(df, values='Period', index=['Data_value'], columns = ['STATUS'])
table

```

```python
table[['F', 'P', 'R']].plot()
```
